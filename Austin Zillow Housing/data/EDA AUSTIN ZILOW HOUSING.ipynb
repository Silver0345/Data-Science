```python
# define numerical & categorical columns
numeric_features = [feature for feature in df.columns if df[feature].dtype != 'O']
categorical_features = [feature for feature in df.columns if df[feature].dtype == 'O']

# print columns
print('We have {} numerical features : {}'.format(len(numeric_features), numeric_features))
print('\nWe have {} categorical features : {}'.format(len(categorical_features), categorical_features))
```

### Dealinng With The Missing Value

```python
housing = df


housing.head(3)
```

### Dealing With The Missing Values
- By dropping 'badgeInfo' column because it has no values
- By filling the missing categorical values with their modes
- By filling the numcial missing values with the means

```python

[housing[cat] == housing[cat].fillna(housing[cat].mode()[0],inplace=True) for cat in categorical_features  ]
[housing[num] == housing[num].fillna(housing[num].mean(),inplace=True) for num in numeric_features  ]
housing = housing.drop(['badgeInfo'], axis=1)
housing.info()
```

```python
plt.subplots(1,3,figsize=(25,6))
plt.subplot(141)
sns.histplot(x='beds', data=housing,kde=True,hue='beds')
plt.subplot(142)
sns.histplot(x='baths',data=housing,kde=True,hue='baths')
plt.subplot(143)
sns.histplot(x='area',data=housing,kde=True)
plt.show()
```

```python
beds = pd.DataFrame(housing['beds'].sort_values(ascending = False))
price = pd.DataFrame(housing['price'].sort_values(ascending = False))
bed_price = beds.join(price)
bed_price.head()
```

```python
plt.figure(figsize=(15,6))
sns.barplot(y='beds', x='price', data= bed_price[:50])
plt.xticks(rotation=90)
plt.title(' Top 50 Number of Bedrooms Vs Price');
```

```python
plt.figure(figsize=(15,6))
sns.barplot(y='beds', x='price', data= bed_price[-50:])
plt.xticks(rotation=90)
plt.title(' Bottom 50 Number of Bedrooms Vs Price');
```

```python
beds_percentage = (housing['beds'].value_counts()/df['beds'].count())*100
beds_percentage
```

```python
keys = ['3 beds','4 beds', '2 beds', '5 beds', '1 bed', '3.33 beds', '6 beds', '0 bed', '7 beds']

palette_color = sns.color_palette('bright')
explode = [0, 0, 0 , 0, 0, 0.1, 0.1, 0.3 , 0.1]
  
plt.figure(figsize=(15,6))
plt.pie(beds_percentage, labels=keys,explode=explode, colors=palette_color, autopct='%.2f%%')
plt.title('Number of Bedroom(s) Percentage');
plt.show()
```
